## 1. Scrape the details of most viewed videos on YouTube from Wikipedia. 
**Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details: 
A) Rank
B) Name
C) Artist
D) Upload date
E) Views**

```python
#Connecting selenium with web browser-
driver= webdriver.Chrome()
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
#Creating variables for required data-
rank=[]
name=[]
artist=[]
upload_date=[]
view=[]
```

```python
#Scraping name from the given page-
name_tag= driver.find_elements(By.XPATH,'//div[@class="mw-body-content"]/div[1]/table[1]/tbody/tr/td[1]')
for i in name_tag:
    names=i.text
    name.append(names)
            
#Scraping artist from the given page-
artist_tag= driver.find_elements(By.XPATH,'//div[@class="mw-body-content"]/div[1]/table[1]/tbody/tr/td[2]')
for i in artist_tag:
    artists=i.text
    artist.append(artists)

#Scraping upload date from the given page-
upload_date_tag= driver.find_elements(By.XPATH,'//div[@class="mw-body-content"]/div[1]/table[1]/tbody/tr/td[4]')
for i in upload_date_tag:
    uploads=i.text
    upload_date.append(uploads)
            
#Scraping views from the given page-
view_tag= driver.find_elements(By.XPATH,'//div[@class="mw-body-content"]/div[1]/table[1]/tbody/tr/td[3]')
for i in view_tag:
    views=i.text
    view.append(views)
```

```python
print(len(name),len(artist),len(upload_date),len(view))
```

```python
rank=list(range(1,31))
```

```python
#Creating DataFrame for the required data-
df=pd.DataFrame({'Rank':rank,'Name':name,'Artist':artist,'Upload Date':upload_date,'Views (Billions)':view})
df
```

```python
df['Name']=df['Name'].str.split('[').str.get(0).str.replace('"',"")
df
```

## 4. Scrape the details of trending repositories on Github.com.
**Url = https://github.com/
You have to find the following details:
A) Repository title
B) Repository description
C) Contributors count
D) Language used**

**Note: - From the home page you have to click on the trending option from Explore menu through code.**

```python
#Connecting selenium with web browser-
driver= webdriver.Chrome()
driver.get('https://github.com/')
```

```python
#Navigating to Open Source > Trending-
open_source=driver.find_element(By.XPATH,"/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/button")
open_source.click()

trending=driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/div/div[3]/ul/li[2]/a')
trending.click()
```

```python
#Creating variables for required data-
title=[]
description=[]
contributors_count=[]
language_used=[]
```

```python
#Scraping repository title from the given page-
try:
    title_tag= driver.find_elements(By.XPATH,'//h2[@class="h3 lh-condensed"]/a')
    for i in title_tag:
        titles=i.text
        title.append(titles)
except: 
    NoSuchElementException
    title.append('-')
            
#Scraping repository description from the given page-
try:
    description_tag= driver.find_elements(By.XPATH,'//p[@class="col-9 color-fg-muted my-1 pr-4"]')
    for i in description_tag:
        descriptions=i.text
        description.append(descriptions)
except:
    NoSuchElementException
    description.append('-')
    
#Scraping contributors count from the given page-
try:
    contributors_count_tag= driver.find_elements(By.XPATH,'//div[@class="f6 color-fg-muted mt-2"]/a[2]')
    for i in contributors_count_tag:
        contributors=i.text
        contributors_count.append(contributors)
except: 
    NoSuchElementException
    contributors_count.append('-')
            
#Scraping language used from the given page-
try:
    language_used_tag= driver.find_elements(By.XPATH,'//div[@class="f6 color-fg-muted mt-2"]/span[1]')
    for i in language_used_tag:
        languages=i.text
        language_used.append(languages)
except:
    NoSuchElementException
    language_used.append('-')
```

```python
#Scrap all repositories URls-
repositories_urls=[]
start=0
end=1
for repo in range(start,end):
    urls=driver.find_elements(By.XPATH,'//h2[@class="h3 lh-condensed"]/a')
    for i in urls:
        repositories_urls.append(i.get_attribute('href'))
```

```python
#Extracting data as required-
for url in repositories_urls:
    driver.get(url)
    time.sleep(2)
    
    try:
        title_tag= driver.find_elements(By.XPATH,'//div[@class="pt-3 hide-full-screen"]/div[1]/div[1]')
        title.append(title_tag.text)
    except: 
        NoSuchElementException
        title.append('-')
    try:
        description_tag= driver.find_elements(By.XPATH,'//p[@class="f4 my-3"]')
        description.append(description_tag.text)
    except:
        NoSuchElementException
        description.append('-')
    try:
        contributors_count_tag= driver.find_elements(By.XPATH,'//div[@class="BorderGrid about-margin"]/div[5]/div/h2/a/span')
        contributors_count.append(contributors_count_tag.text)
    except: 
        NoSuchElementException
        contributors_count.append('-')
    try:
        language_used_tag= driver.find_elements(By.XPATH,'//div[@class="BorderGrid about-margin"]/div[6]/div/ul')
        language_used.append(language_used_tag.text)
    except:
        NoSuchElementException
        language_used.append('-')
```

```python
print(len(title),len(description),len(contributors_count),len(language_used))
```

```python
df=pd.DataFrame({'Repository title':title,'Contributors count':contributors_count,'Language used':language_used})
df
```

## 7. Scrape the details most watched tv series of all time from imdb.com.
**Url = https://www.imdb.com/list/ls095964455/ You have to find the following details:
A) Name
B) Year span
C) Genre
D) Run time
E) Ratings
F) Votes**

```python
#Connecting selenium with web browser-
driver= webdriver.Chrome()
driver.get('https://www.imdb.com/list/ls512407256/')
```

```python
#Creating variables for required data-
name=[]
year_span=[]
genre=[]
run_time=[]
ratings=[]
votes=[]
```

```python
start=0
end= 1

for page in range(start, end):

#Scraping name from the given page-
    name_tag= driver.find_elements(By.XPATH,'//h3[@class="lister-item-header"]/a')
    for i in name_tag:
        names=i.text
        name.append(names)
            
#Scraping year span from the given page-
    year_span_tag= driver.find_elements(By.XPATH,'//h3[@class="lister-item-header"]/span[2]')
    for i in year_span_tag:
        year_spans=i.text
        year_span.append(year_spans)
    
#Scraping genre from the given page-
    genre_tag= driver.find_elements(By.XPATH,'//p[@class="text-muted text-small"]/span[5]')
    for i in genre_tag:
        genres=i.text
        genre.append(genres)
    
#Scraping run time from the given page-
    run_time_tag= driver.find_elements(By.XPATH,'//p[@class="text-muted text-small"]/span[3]')
    for i in run_time_tag:
        run_times=i.text
        run_time.append(run_times)
            
#Scraping ratings from the given page-
    ratings_tag= driver.find_elements(By.XPATH,'//div[@class="ipl-rating-widget"]/div/span[2]')
    for i in ratings_tag:
        rating=i.text
        ratings.append(rating)
    
#Scraping votes from the given page-
    votes_tag= driver.find_elements(By.XPATH,'//div[@class="lister-item-content"]/p[4]/span[2]')
    for i in votes_tag:
        vote=i.text
        votes.append(vote)
    
#Next button
    next_button= driver.find_element(By.XPATH,'/html/body/div[2]/div/div[2]/div/div[1]/div/div[3]/div[5]/div/div/div/a[2]')
    next_button.click()
    time.sleep(3)
```

```python
print(len(name),len(year_span),len(genre),len(run_time),len(ratings),len(votes))
```

```python
df=pd.DataFrame({'Name':name,'Year span':year_span,'Genre':genre,'Run time':run_time,'Ratings':ratings,'Votes':votes})
df
```

## 8. Details of Datasets from UCI machine learning repositories.
**Url = https://archive.ics.uci.edu/ You have to find the following details:
A) Dataset name
B) Data type
C) Task
D) Attribute type
E) No of instances
F) No of attribute 
G) Year**

**Note: - from the home page you have to go to the Show All Dataset page through code.**

```python
#Connecting selenium with web browser-
driver= webdriver.Chrome()
driver.get('https://archive.ics.uci.edu/')
```

```python
#Navigating to all datasets page:
view_datasets=driver.find_element(By.XPATH,'//div[@class="flex flex-wrap justify-center gap-5"]/a[1]')
view_datasets.click()
```

```python
#Scrap all products URls-
dataset_urls=[]
start=0
end=67
for page in range(start,end):
    urls=driver.find_elements(By.XPATH,'//h2[@class="truncate text-primary"]/a')
    for i in urls:
        dataset_urls.append(i.get_attribute('href'))
    next_button=driver.find_element(By.XPATH,'//div[@class="btn-group"]/button[2]')
    next_button.click()
    time.sleep(2)
```

```python
len(dataset_urls)
```

```python
#Creating variables for required data-
dataset_name=[]
data_type=[]
task=[]
attribute_type=[]
no_of_instances=[]
no_of_attribute=[]
year=[]
```

```python
#Extracting data as required-
for url in dataset_urls:
    driver.get(url)
    time.sleep(2)
    
    try:
        name= driver.find_element(By.XPATH,'//div[@class="relative flex flex w-full items-center gap-4 bg-primary p-2"]/div[2]/div/h1')
        dataset_name.append(name.text)
    except: 
        NoSuchElementException
        dataset_name.append('-')
    try:
        data_type_tag= driver.find_element(By.XPATH,'//div[@class="col-span-4"][1]/p')
        data_type.append(data_type_tag.text)
    except:
        NoSuchElementException
        data_type.append('-')
    try:
        tasks= driver.find_element(By.XPATH,'//div[@class="col-span-4"][3]/p')
        task.append(tasks.text)
    except: 
        NoSuchElementException
        task.append('-')
    try:
        attribute_types= driver.find_element(By.XPATH,'//div[@class="col-span-4"][4]/p')
        attribute_type.append(attribute_types.text)
    except:
        NoSuchElementException
        attribute_type.append('-')
    try:
        no_of_instance= driver.find_element(By.XPATH,'//div[@class="col-span-4"][5]/p')
        no_of_instances.append(no_of_instance.text)
    except:
        NoSuchElementException
        no_of_instances.append('-')
    try:
        no_of_attributes= driver.find_element(By.XPATH,'//div[@class="col-span-4"][6]/p')
        no_of_attribute.append(no_of_attributes.text)
    except: 
        NoSuchElementException
        no_of_attribute.append('-')
    try:
        years= driver.find_element(By.XPATH,'//h2[@class="text-sm text-primary-content"]')
        year.append(years.text)
    except: 
        NoSuchElementException
        year.append('-')
```

```python
print(len(dataset_name),len(data_type),len(task),len(attribute_type),len(no_of_instances),len(no_of_attribute),len(year))
```

```python
df=pd.DataFrame({'Dataset name':dataset_name,'Data type':data_type,'Attribute type':attribute_type,'No. of instances':no_of_instances,'No. of attributes':no_of_attribute,'Year':year})
df
```

```python
df['Year']=df['Year'].str.split('on ').str.get(1)
df
```

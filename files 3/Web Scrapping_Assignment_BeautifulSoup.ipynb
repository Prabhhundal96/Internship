### 1) Write a python program to display IMDB’s Top rated 100 Indian movies’ data https://www.imdb.com/list/ls056092300/ (i.e. name, rating, year ofrelease) and make data frame.

```python
#Get request sent to website-
page=requests.get('https://www.imdb.com/list/ls056092300/')
page
```

```python
#BeautifulSoup object created- 
Soup=BeautifulSoup(page.content)
```

```python
#Extracting movie name and year of release-
movie_details=[]
for i in Soup.find_all('h3',class_="lister-item-header"):
    movie_details.append(i.text)
movie_details
```

```python
#Extracting movie ratings-
ratings=[]
for i in Soup.find_all('div',class_="ipl-rating-star small"):
    ratings.append(i.text)
ratings
```

```python
#Creating dataframe-
df=pd.DataFrame({'Movie_details':movie_details,'Ratings':ratings})
df
```

```python
#Cleaning data-
df['Movie_details']=df['Movie_details'].str.replace('\n+','',regex=True)
df['Ratings']=df['Ratings'].str.replace('\n+','',regex=True)
df
```

```python
df['Movie_name']=df['Movie_details'].str.split('(').str.get(0).str.replace('\d.','',regex=True)
df['Year']=df['Movie_details'].str.split('(').str.get(1).str.replace(')','',regex=True)
df
```

```python
df=df.drop('Movie_details',axis=1)
```

```python
c1=df['Movie_name']
df.insert(0,'Name',c1)
```

### 6) Write a python program to scrape first 10 product details which include product name , price , Image URL from https://www.bewakoof.com/bestseller?sort=popular .

```python
#Get request sent to website
page6=requests.get('https://www.bewakoof.com/bestseller?sort=popular')
page6
```

```python
#Creating BeautifulSoup object
soup6=BeautifulSoup(page6.content)
```

```python
#Exracting product name-
p_name=[]
for i in soup6.find_all('h2',class_="clr-shade4 h3-p-name undefined false")[:10]:
    p_name.append(i.text)
p_name
```

```python
#Exracting product price-
price=[]
for i in soup6.find_all('div',class_="discountedPriceText clr-p-black false")[:10]:
    price.append(i.text)
price
```

```python
#Exracting image url-
image_url=[]
for i in soup6.find_all('img',class_="productImgTag")[:10]:
    image_url.append(i['src'])
image_url
```

```python
#Creating Dataframe-
df6=pd.DataFrame({'Product name':p_name,'Price':price,'Image URL':image_url})
df6
```

### 7) Please visit https://www.cnbc.com/world/?region=world and scrap- a) headings b) date c) News link

```python
#Get request sent to website-
page7=requests.get('https://www.cnbc.com/world/?region=world')
page7
```

```python
#Creating BeautifulSoup object-
soup7=BeautifulSoup(page7.content)
```

```python
# Extracting news hadings-
headings = []
for i in soup7.find_all('a',class_="Card-title"):
    headings.append(i.text)
headings
```

```python
# Extracting news link-
link = []
for i in soup7.find_all('a',class_="Card-title"):
    link.append(i['href'])
link
```

```python
# Extracting news date-
date = []
for i in link:
    x=re.findall('\d{4}/\d{2}/\d{2}',i)
    date.append(x)
date
```

```python
#Creating dataframe-
df7=pd.DataFrame({'News heading':headings,'News Date':date,'News link':link})
df7
```

### 8) Please visit https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloaded- articles/ and scrap- a) Paper title b) date c) Author

```python
#Get request sent to website-
page8=requests.get('https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloaded-articles/')
page8
```

```python
#Create BeautifulSoup Object-
soup8=BeautifulSoup(page8.content)
```

```python
#Extracting Paper Title-
title=[]
for i in soup8.find_all('h2',class_="h5 article-title"):
    title.append(i.text.strip())
title
```

```python
#Extracting Date-
date=[]
for i in soup8.find_all('p',class_="article-date"):
    date.append(i.text)
date
```

```python
#Extracting Paper Author-
author=[]
for i in soup8.find_all('p',class_="article-authors"):
    author.append(i.text.strip())
author
```

```python
#Creating dataframe-
df8=pd.DataFrame({'Paper_title':title,'Date':date,'Author':author})
df8
```

```python
#Replacing "|" in author name with ","-
df8['Author']=df8['Author'].str.replace('|',',',regex=True)
```

5. Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps.

```python
#Connecting selenium with web browser-
driver=webdriver.Chrome()
driver.get('https://www.google.com/maps')
```

```python
#Function to get latitude and longitude of a city
def get_city_coordinates(city):
    # find and clear the search box
    search_box = driver.find_element(By.XPATH,'/html/body/div[1]/div[3]/div[8]/div[3]/div[1]/div[1]/div/div[2]/form/input')
    search_box.clear()
    
    # enter the city name in the search box
    search_box.send_keys(city)
    
    # submit the search query by pressing Enter key
    search=driver.find_element(By.XPATH,'/html/body/div[1]/div[3]/div[8]/div[3]/div[1]/div[1]/div/div[2]/div[1]/button/span')
    search.click()
    
    try:
        # wait for the map to load with timeout of 10 seconds 
        driver.implicitly_wait(10)
        
        # find and extract latitude and longitude from URL parameter 'sll'
        url = driver.current_url
        
        start_index = url.index('@') + 1   # find starting index after '@' symbol in URL 
        end_index = url.index('z') - 1     # find ending index before 'z' symbol in URL
        
        coordinates_str = url[start_index:end_index]
        
        latitude,longitude = coordinates_str.split(',')
        
        return float(latitude), float(longitude)
    
    except Exception as e:
        print("Error occurred while getting coordinates:", str(e))
```

```python
#Test by searching for New York City
city_name = "New York City"
latitude,longitude = get_city_coordinates(city_name)

print(f"Coordinates of {city_name}: Latitude - {latitude}, Longitude - {longitude}")
```
